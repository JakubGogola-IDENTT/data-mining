```python
import pandas as pd
from statsmodels.formula.api import ols
```

# Exercise 11

```python
auto_df = pd.read_csv('./data/Auto.csv')
```

```python
auto_df
```

```python
auto_df = auto_df[auto_df['horsepower'] != '?']
auto_df = auto_df.astype({
    'horsepower': float
})
```

# Year variable

## I model

Let's use `year` as continuous variable.

```python
model_1 = ols(formula='mpg ~ C(year)', data=auto_df).fit()
model_1.summary()
```

## Model II

Let's use `year` as categorical variable.

```python
model_2 = ols(formula='mpg ~ year', data=auto_df).fit()
model_2.summary()
```

## Q&A

### Which model is better?
The second one which treats year as categorical variable.

### What if there were more than 13 values for variable `year`?
In the case where variable is continuous - nothing will change. In the case where variable is categorical - there'll be more parameters to train

### Which of those two models is easier to train?
Continuous model because it has less parameters to train.
